```python
import time
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
data1=[]

with Chrome(executable_path=r'C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe') as driver:
    wait = WebDriverWait(driver,20)
    driver.get("https://youtu.be/UqKVL56IJB8")

    for item in range(100): 
        wait.until(EC.visibility_of_element_located((By.TAG_NAME,"body"))).send_keys(Keys.END)
        time.sleep(10)
        if (wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#content-text")))):
            for comment in wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#content-text"))):
                if (comment.text not in data1 and len(data1)<100):
                    data1.append(comment.text)
        if (len(data1) >= 100):
            break

df1 = pd.DataFrame(data1, columns=['comment'])
df1.head()
```
